```python
import pandas as pd 
import numpy as np
```

```python
df=pd.DataFrame({ 'product':['carrots','braccoli','banna','banna',
                            'beans','orange','braccoli','banna'],
                 'categary':['vegetable','vegetavle','fruit','fruit','vegetable','vegetavle','fruit','fruit'],
                 'quantity':[8,5,3,4,5,9,11,8],
                 'amount':[270,239,617,384,626,610,62,90]
    
})
```

```python
df
```

```python
result=pd.pivot_table(df,index=['product','categary'],values=["amount"],aggfunc=[np.sum])
```

```python
result
```

```python
result=pd.pivot_table(df,index=['product'],values=['amount'],aggfunc=[np.min,np.median,np.mean])
```
